```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# read in highway map
hw_map_df = pd.read_csv('./data/highway_map.csv', header=None, sep=' ')
hw_map_df.head()
```

```python
hw_map = hw_map_df.values
```

```python
hw_map.shape
```

```python
fig = plt.figure(figsize=(12,9))
fig.suptitle('Highway Map', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)

ax.plot(hw_map[:,0], hw_map[:,1], '-o')

ax.plot(hw_map[0,0], hw_map[0,1], 'ro')
ax.annotate('Starting Point', xy=(hw_map[0,0], hw_map[0,1]), xytext=(hw_map[0,0]+200, hw_map[0,1]+200),
            arrowprops=dict(facecolor='black', shrink=0.05))

for p in hw_map:
    ax.arrow(p[0], p[1], p[3]*100, p[4]*100, head_width=20, head_length=50, fc='k', ec='k')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,2500)
ax.set_ylim(900, 3200)
```
